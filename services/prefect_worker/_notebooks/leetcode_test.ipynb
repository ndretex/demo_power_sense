```python
from itertools import combinations
from typing import List
class Solution:
    def __init__(self):
        pass

    def check_if_sum_equals_target(self, ii, comb, target):
        if sum(comb)==target:
            return self.combs_index[ii]
        return []

    def twoSum(self, nums: List[int], target: int) -> List[int]:
        self.combs = list(combinations(nums, 2))
        self.combs_index = list(combinations(range(len(nums)),2))
        index_sum_match_target = [self.check_if_sum_equals_target(ii, comb, target) for ii, comb in enumerate(self.combs)]
        flat_indexes = [item for sublist in index_sum_match_target for item in sublist]

        return flat_indexes

        
```

```python
nums = [2,7,11,15]
sol = Solution()
result = sol.twoSum(nums, 9)
print("result:")
print(result)
```
